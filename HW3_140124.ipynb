
linear regression Лінійна регресія

$$h(x)=w_0+w_1*x$$


h($\vec x$)=$\vec w$⋅$\vec x$




```python
def hypothesis(w, X):

    return np.dot(X, w)
```

Функція втрат

```python
def cost_func(w, X, y):

    m = len(y)
    h = hypothesis(w, X)
    error = h - y
    return np.sum(error ** 2) / (2 * m)
```

```python
def gradient_descent_step(w, X, y, learning_rate):

    m = len(y)
    h = hypothesis(w, X)
    error = h - y
    gradient = np.dot(X.T, error) / m
    w = w - learning_rate * gradient
    return w
```

```python
features = ['area']
target = 'price'
X = df_house[features]
y = df_house[target]

model = LinearRegression()
model.fit(X, y)

print(f'Зсув: {model.intercept_}')
print(f'Нахил: {model.coef_}')
```

```python
X = np.hstack((np.ones((X.shape[0], 1)), X))
w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(f'Зсув: {w[0]}')
print(f'Нахил: {w[1:]}')
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Згенеруємо значення для графіку
x_values = np.linspace(min(X['area']), max(X['area']), 100)
bathrooms_values = np.linspace(min(X['bathrooms']), max(X['bathrooms']), 100)
bedrooms_values = np.linspace(min(X['bedrooms']), max(X['bedrooms']), 100)

# Створимо комбінації значень
x, bathrooms, bedrooms = np.meshgrid(x_values, bathrooms_values, bedrooms_values)

# Розрахунок відповідних значень 'price' для кожної комбінації
y_values = model1.intercept_ + model1.coef_[0] * x + model1.coef_[1] * bathrooms + model1.coef_[2] * bedrooms

# Побудуємо 3D графік
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Візуалізація
ax.scatter(X['area'], X['bathrooms'], y, color='blue', label='Actual data')  # Реальні дані
ax.plot_trisurf(x.flatten(), bathrooms.flatten(), y_values.flatten(), color='red', alpha=0.5)  # Лінійна регресія

ax.set_xlabel('Area')
ax.set_ylabel('Bathrooms')
ax.set_zlabel('Price')
ax.set_title('Linear Regression with Bedrooms and Bathrooms')
plt.show()


```
